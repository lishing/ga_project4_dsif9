## Combining Dengue Cases, Google Trends & Dengue Cases

### Load the libraries

```python
import pandas as pd
```

### Reading in the Datasets

```python
# Read in the Cleaned Changi Weather dataset
weekly_climate_dengue=pd.read_csv('../../assets/cleaned_comb_changi_weather.csv', parse_dates=['date'], index_col='date')
```

### Weekly Climate & Dengue

```python
weekly_climate_dengue.head()
```

```python
# Dropping Year, Month & Day columns
weekly_climate_dengue.drop(columns=['year','month','Year','Month','Day','week'], inplace=True)
```

```python
# Keeping the Dengue cases only
weekly_climate_dengue=weekly_climate_dengue[weekly_climate_dengue['type_dengue']=='Dengue']
```

```python
weekly_climate_dengue.head()
```

```python
# Drop the type_dengue column
weekly_climate_dengue.drop(columns=['type_dengue'], inplace=True)
```

```python
weekly_climate_dengue.shape
```

### Weekly Google Trends

```python
# Read in the Complete Google Trends dataset
google=pd.read_csv('../../assets/complete_google_trends.csv')
```

```python
google.head()
```

```python
# Drop unnamed: 0 column from google
google.drop(columns=['Unnamed: 0', 'year','month','week'], inplace=True)
```

```python
# convert date column to datetime format
google['date'] = pd.to_datetime(google['date'], format='%Y-%d-%m')
```

```python
google.head()
```

```python
# set the date column as the index
google.set_index('date', inplace=True)
```

```python
google.head()
```

```python
google.shape
```

### Concatenate both dataframes

```python
# concatenate the dataframes on the same time index
dengue_weather_google = pd.merge(weekly_climate_dengue, google, left_index=True, right_index=True)
```

```python
dengue_weather_google.head()
```

```python
dengue_weather_google
```

### Exporting out the combined Dengue, Weather and Google Trends

```python
dengue_weather_google.to_csv('../../assets/dengue_weather_google.csv')
```
