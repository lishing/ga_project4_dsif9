# Collecting Google trends for period 2019

In this section, we collect the Google trends for the period of 2019. The data collected will act as part of the exogenous variables for the SARIMA forecast of the dengue cases for the 'unseen' 2019.

```python
import pandas as pd
```

```python
# Reading in all the individual google trends for year 2019
fog=pd.read_csv('../../assets/fogging_search.csv')
den_fever=pd.read_csv('../../assets/dengue_fever_search.csv')
dengue=pd.read_csv('../../assets/dengue_search.csv')
den_symp=pd.read_csv('../../assets/dengue_symptoms_search.csv')
fever=pd.read_csv('../../assets/fever_search.csv')
headache=pd.read_csv('../../assets/headache_search.csv')
insect=pd.read_csv('../../assets/insect_repellent_search.csv')
mosquito=pd.read_csv('../../assets/mosquito_search.csv')
nausea=pd.read_csv('../../assets/nausea_search.csv')
rashes=pd.read_csv('../../assets/rashes_search.csv')
vomit=pd.read_csv('../../assets/vomiting_search.csv')
wolbachia=pd.read_csv('../../assets/wolbachia_search.csv')
eye_pain=pd.read_csv('../../assets/eye_pain_search.csv')
joint_pain=pd.read_csv('../../assets/joint_pain_search.csv')
```

```python
def add_day (df):
    '''
    Input a google trend dataframe.
    Output a dataframe where Monday is the starting day of the week and date is set as index.
    '''
    df['date'] = pd.to_datetime(df['date']) + pd.Timedelta(days=1)
    return df
```

```python
# combining all trends into one combined dataframe
df_combined=pd.merge(fog,den_fever, on='date')
df_combined2=pd.merge(df_combined, dengue, on='date')
df_combined3=pd.merge(df_combined2, den_symp, on='date')
df_combined4=pd.merge(df_combined3, fever, on='date')
df_combined5=pd.merge(df_combined4, headache, on='date')
df_combined6=pd.merge(df_combined5, insect, on='date')
df_combined7=pd.merge(df_combined6, mosquito, on='date')
df_combined8=pd.merge(df_combined7, nausea, on='date')
df_combined9=pd.merge(df_combined8, rashes, on='date')
df_combined10=pd.merge(df_combined9, vomit, on='date')
df_combined11=pd.merge(df_combined10, wolbachia, on='date')
df_combined12=pd.merge(df_combined11, eye_pain, on='date')
df_combined13=pd.merge(df_combined12, joint_pain, on='date')
```
